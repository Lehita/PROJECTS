### DATA PRE-PROCESSING

```python
data.head()
```

```python
#converting the gender and risk level columns to integers
data["Gender"].replace("Male",0,inplace=True)
data["Gender"].replace("Female",1,inplace=True)
data["Level"].replace("Low",0,inplace=True)
data["Level"].replace("Medium",1,inplace=True)
data["Level"].replace("High",2,inplace=True)
data["Genetic_risk"].replace("yes",1,inplace=True)
data["Genetic_risk"].replace("no",0,inplace=True)
```

```python
data.head()
```

### FEATURE ENGINEERING

```python
#visualizing correlation coefficient of the columns.
corr = data.corr()
```

```python
plt.figure(figsize=(14,6))
sns.heatmap(data.corr(),annot=True,cmap='coolwarm')
plt.show()
```

```python
#feature selection
x = data[data.columns[0:len(data.columns)-1]]
y = data["Level"]
from sklearn.ensemble import ExtraTreesClassifier
tree = ExtraTreesClassifier(random_state=0)
tree.fit(x,y)
print(tree.feature_importances_)
```

```python
#visualizing the 15 most important features (columns)
feat_importance = pd.Series(tree.feature_importances_,index=x.columns)
feat_importance.nlargest(15).plot(kind='barh')
col_of_interest = feat_importance.nlargest(15).index.tolist()
col_of_interest
```

```python
#splitting the data
features = data[col_of_interest]
target = data["Level"]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.20,shuffle=False,random_state=0)
```

```python
#reducing the dimensionality of the data
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
print(pca.explained_variance_ratio_)
```

### MODELLING

```python
#using Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_validate
clf1 = GaussianNB()
scores = cross_validate(clf1,x_train,y_train,cv=4)
scores["test_score"].mean()
```

```python
#using SVC
from sklearn.svm import SVC
clf2 = SVC()
scores2 = cross_validate(clf2,x_train,y_train,cv=4)
scores2["test_score"].mean()
```

```python
#using DecisionTree classifier
from sklearn.tree import DecisionTreeClassifier
clf3 = DecisionTreeClassifier(max_depth=3)
scores3 = cross_validate(clf3,x_train,y_train,cv=4)
scores3["test_score"].mean()
```

```python
#using RandomForest classifier
from sklearn.ensemble import RandomForestClassifier
clf4 = RandomForestClassifier(max_depth=3,random_state=0)
scores4 = cross_validate(clf4,x_train,y_train,cv=4)
scores4["test_score"].mean()
```

```python
#using Logistic regression model
from sklearn.linear_model import LogisticRegression
clf5 = LogisticRegression()
scores5 = cross_validate(clf5,x_train,y_train,cv=4)
scores5["test_score"].mean()
```

#### Choosing RandomForest classifier

```python
import sklearn
clf = sklearn.base.clone(clf4) #cloning the randomforest classifier trained above
clf.fit(x_train,y_train) #training the classifier using training + validation datasets
#checking the accuracy
from sklearn.metrics import confusion_matrix, accuracy_score
pred = clf.predict(x_test) #predicting the test set
cm = confusion_matrix(y_test, pred)
print(cm)
accuracy_score(y_test, pred)
```

```python
import pickle as pkl
#saving the model as a pickle file
pickle_out1 = open("clf.pkl","wb")
pkl.dump(clf,pickle_out1)
pickle_out1.close()
#saving the pca processor as a pickle file
pickle_out2 = open("pca.pkl","wb")
pkl.dump(pca,pickle_out2)
pickle_out2.close()
```
