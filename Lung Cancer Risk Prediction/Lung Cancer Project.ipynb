```python
#converting the gender and risk level columns to integers
data["Gender"].replace("Male",1,inplace=True)
data["Gender"].replace("Female",2,inplace=True)
data["Level"].replace("Low",0,inplace=True)
data["Level"].replace("Medium",1,inplace=True)
data["Level"].replace("High",2,inplace=True)
```

### FEATURE ENGINEERING

```python
#using heatmaps
corr = data.corr()
```

#### Multicollinearity can be seen in the heatmap. Thus, I will not be using the heatmap to select features.

```python
#using Extra trees classifier
x = data[data.columns[0:len(data.columns)-1]]
y = data["Level"]
from sklearn.ensemble import ExtraTreesClassifier
tree = ExtraTreesClassifier()
tree.fit(x,y)
print(tree.feature_importances_)
```

```python
feat_importance = pd.Series(tree.feature_importances_,index=x.columns)
feat_importance.nlargest(15).plot(kind='barh')
col_of_interest = feat_importance.nlargest(15).index.tolist()
col_of_interest
```

```python
#using SelectKBest
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
select = SelectKBest(score_func=chi2,k=15)
fit = select.fit_transform(x,y)
filters = x.columns[select.get_support()].to_list()
filters
```

### MODELLING

```python
#splitting the data
features = data[col_of_interest]
target = data["Level"]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.20,shuffle=False,random_state=0)
```

```python
#reducing the dimensionality of the data
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
print(pca.explained_variance_ratio_)
```

```python
#training the model using Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_validate
clf1 = GaussianNB()
scores = cross_validate(clf1,x_train,y_train,cv=4)
scores["test_score"].mean()
```

```python
#training the model using SVC
from sklearn.svm import SVC
clf2 = SVC()
scores2 = cross_validate(clf2,x_train,y_train,cv=4)
scores2["test_score"].mean()
```

```python
#training the model using DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
clf3 = DecisionTreeClassifier(max_depth=3)
scores3 = cross_validate(clf3,x_train,y_train,cv=4)
scores3["test_score"].mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
clf4 = RandomForestClassifier(max_depth=3, random_state=0)
scores4 = cross_validate(clf4,x_train,y_train,cv=4)
scores4["test_score"].mean()
```

```python
from sklearn.linear_model import LogisticRegression
clf5 = LogisticRegression()
scores5 = cross_validate(clf5,x_train,y_train,cv=4)
scores5["test_score"].mean()
```

```python
#### choosing RandomForest classifier
```

```python
import sklearn
clf = sklearn.base.clone(clf4)
clf.fit(x_train,y_train)
#checking the accuracy
from sklearn.metrics import confusion_matrix, accuracy_score
pred = clf.predict(x_test)
cm = confusion_matrix(y_test, pred)
print(cm)
accuracy_score(y_test, pred)
```
